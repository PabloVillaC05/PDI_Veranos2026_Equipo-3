**Importar librerías**

```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files
```

**Agregando imagen desde archivos**

```python
uploaded = files.upload()
```

**Leer imagen**

```python
imagen = cv2.imread("ave.jpg")
```

**De BGR a RGB**

```python
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
```

**Escala de grises**

```python
gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
```

**Histograma con OpenCV**



```python
hist = cv2.calcHist([gris], [0], None, [256], [0, 256])
```

**Mostrar imagenes**

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.imshow(imagen_rgb)
plt.title("Imagen original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(gris, cmap="gray")
plt.title("Imagen en escala de grises")
plt.axis("off")

plt.subplot(1,3,3)
plt.plot(hist,color='maroon')
plt.title("Histograma usando OpenCV")
plt.xlabel("Escala de grises")
plt.ylabel("Número de píxeles")

plt.tight_layout()
plt.show()
```

**Histograma con matplotlib**

```python
plt.figure(figsize=(7,5))

plt.hist(gris.ravel(), bins=256, range=[0,256], color='maroon')

plt.title("Histograma usando matplotlib")
plt.xlabel("Escala de grises")
plt.ylabel("Número de píxeles")

plt.show()
```
